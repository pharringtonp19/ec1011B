### **Libraries**

```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

### **Parameters**

```python
K0 = 20.0
L0 = 20.0
a = 5.0
n = 0.15
d = .1
s = .25
n_periods = 100
```

### **Functions**

```python
def Output(x1, x2):
  return x1**(0.5)* x2**(0.5)

def CapitalStock(a, k, l, d, s):
  investment = s * Output(k, a*l)
  return investment + (1-d)*k

def LaborStock(n, l):
  return (1+n)*l
```

###

```python
capital_trajectory = []
labor_trajectory = []
K = K0
L = L0
for i in range(n_periods):
  K = CapitalStock(a, K, L, d, s)
  L = LaborStock(n, L)
  capital_trajectory.append(K)
  labor_trajectory.append(L)
```

### **Plot: Capital & Labor Trajectories**

```python
fig, ax = plt.subplots(figsize=(7, 4.5))
plt.plot(range(n_periods),capital_trajectory)
plt.plot(range(n_periods),labor_trajectory)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='y', color='lightgray', linestyle='-', linewidth=0.7, alpha=0.6)
ax.set_axisbelow(True)
plt.title('Units', loc='left')
plt.xlabel('Time', size=14)
plt.show()
```

### **Plot: Capital Per Labor Trajectories**

```python
fig, ax = plt.subplots(figsize=(7, 4.5))
plt.plot(range(n_periods),jnp.array(capital_trajectory)/jnp.array(labor_trajectory))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='y', color='lightgray', linestyle='-', linewidth=0.7, alpha=0.6)
ax.set_axisbelow(True)
plt.title('Capital per Labor', loc='left')
plt.xlabel('Time', size=12)
plt.show()
```
